```python
import numpy as np
import pandas as pd
from plotnine import *

foodiex = pd.read_csv('2020-XTern-DS.csv')
latitudemedian = foodiex['Latitude'].median()
longitudemedian = foodiex['Longitude'].median()
plot = ggplot(foodiex, aes(x = 'Latitude', y = 'Longitude')) + geom_boxplot()
plot2 = ggplot(foodiex, aes(x = 'Longitude', y = 'Latitude')) + geom_boxplot()
latq1 = foodiex['Latitude'].quantile(.25)
latq3 = foodiex['Latitude'].quantile(.75)
foodiex['latdata_within_quartiles'] = foodiex['Latitude'].apply(lambda x: True if x<latq3 and x>latq1 else False)
print('Latitude Quartiles')
print(latq1, latq3)
lonq1 = foodiex['Longitude'].quantile(.25)
lonq3 = foodiex['Longitude'].quantile(.75)
foodiex['londata_within_quartiles'] = foodiex['Longitude'].apply(lambda x: True if x<latq3 and x>latq1 else False)
print('Longitude Quartiles')
print(lonq1, lonq3)
latq1 = foodiex['Latitude'].quantile(.25)
latq3 = foodiex['Latitude'].quantile(.75)
foodiex['data_within_quartiles'] = np.where(foodiex['latdata_within_quartiles'] & foodiex['londata_within_quartiles'], True, False)
print('\n All of the Restaurants with a Latitude and Longitude with the Quartiles')
print(foodiex[foodiex['data_within_quartiles'] == True])
print('\n Conclusion: There is no data within both Quartiles')
print(plot)
print(plot2)
foodiex
```
