```python
import numpy as np
import pandas as pd
from plotnine import *

foodiex = pd.read_csv('2020-XTern-DS.csv')
foodiex['Average_Cost'] = foodiex['Average_Cost'].apply(lambda x: x.strip('$'))
foodiex['Average_Cost'] = foodiex['Average_Cost'].apply(lambda x: x.replace(',', ''))
foodiex['Average_Cost'] = foodiex['Average_Cost'].apply(lambda x: float(x))
foodiex = foodiex.replace('-', np.nan)
foodiex = foodiex.replace('NEW', np.nan)
foodiex = foodiex.dropna(how='any')
foodiex['Rating'] = foodiex['Rating'].apply(lambda x: float(x))
lowcost_bestrating = foodiex[(foodiex['Average_Cost'] > 25) & (foodiex['Rating'] < 3.0)]
plot = (ggplot(foodiex, aes(x = 'Rating', y = 'Average_Cost')) + geom_point(fill = 'lightblue'))
print(plot)
lowcost_bestrating
```
