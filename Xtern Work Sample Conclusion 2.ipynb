```python
import numpy as np
import pandas as pd
from plotnine import *

foodiex = pd.read_csv('2020-XTern-DS.csv')
y = ','
foodiex['maincuisine'] = foodiex['Cuisines'].apply(lambda x: x if y not in x else x[0:x.index(",")])
foodiex['Cook_Time'] = foodiex['Cook_Time'].apply(lambda x: x[0:x.index(' ')])
foodiex['Cook_Time'] = foodiex['Cook_Time'].apply(lambda x: int(x))
foodiex['counts'] = foodiex.groupby('maincuisine')['maincuisine'].transform('count')
cooktimecuisine = foodiex.groupby('maincuisine').agg('mean')
newcookcuisine = cooktimecuisine.sort_values(by = ['Cook_Time'])
newcookcuisine = newcookcuisine[newcookcuisine['counts'] > 5]
newcookcuisine = newcookcuisine.drop(columns = ['Latitude', 'Longitude'])
print('Least amount of Cook Time among Cuisines with more than 5 restaurants')
print(newcookcuisine.head(10))
print('\n')
print('Most amount of Cook Time among Cuisines with more than 5 restaurants')
print(newcookcuisine.tail(10))
newcookcuisine
```
